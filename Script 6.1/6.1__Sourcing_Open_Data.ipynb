#### 3.0 Data Wrangling

```python
#checking fot mix data type
for col in df_US_Ecom.columns.tolist():
  weird = (df_US_Ecom[[col]].applymap(type) != df_US_Ecom[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_US_Ecom[weird]) > 0:
    print (col)
    
```
